```python
import numpy as np
from numpy import random

length = 20

choices = [1,1,0,0,0]
world = random.choice(choices, size=(length,length))
path = np.zeros((length,length),dtype=np.int32)

a_x, a_y = (0,0) #agent position
t_x, t_y = (random.randint(length/2+1,length),random.randint(length/2+1,length)) # target position

world[t_x][t_y] = 2
world[a_x][a_y] = 0

# 1 are potholes walls, any cell outside the world is inaccesible an throws you back with -2 reward

state = [1,1,world[0][1],world[1][1],world[1][0],1,1,1,a_x,a_y]
# knows every cell around it, even diagonal ones, and its own position

G = 0

def env(action,state):

    a_x = state[8]
    a_y = state[9]
    
    path[a_x][a_y] += 1

    new_state = state[:]

    reward = -1
    
    if action == "up":
        if a_x == 0:
            return reward, new_state, 

        a_x += -1

    elif action == "down":
        if a_x == length-1:
            return reward, new_state, 

        a_x += 1
    
    elif action == "right":
        if a_y == length-1:
            return reward, new_state, 

        a_y += 1
    
    elif action == "left":
        if a_y == 0:
            return reward, new_state, 

        a_y += -1

    new_state[8] = a_x
    new_state[9] = a_y
    
    
    if a_x>0:
        new_state[0] = world[a_x-1][a_y]
    else:
        new_state[0] = 1

    if a_x>0 and a_y<length-1:
        new_state[1] = world[a_x-1][a_y+1]
    else:
        new_state[1] = 1

    if a_y<length-1:
        new_state[2] = world[a_x][a_y+1]
    else:
        new_state[2] = 1
    
    if a_x<length-1 and a_y<length-1:
        new_state[3] = world[a_x+1][a_y+1]
    else:
        new_state[3] = 1
    
    if a_x<length-1:
        new_state[4] = world[a_x+1][a_y]
    else:
        new_state[4] = 1

    if a_x<length-1 and a_y>0:
        new_state[5] = world[a_x+1][a_y-1]
    else:
        new_state[5] = 1

    if a_y>0:    
        new_state[6] = world[a_x][a_y-1]
    else:
        new_state[6] = 1

    if a_x>0 and a_y>0:
        new_state[7] = world[a_x-1][a_y-1]
    else:
        new_state[7] = 1


    if world[a_x][a_y] == 1:
        reward -= 4
    elif world[a_x][a_y]== 2:
        reward += 10

    return reward, new_state, 
```
